```python
questions = {1: 'What is the capital city of France?',
             2: 'Which planet is known as the Red Planet?',
             3: 'How many legs does a spider typically have?',
             4: 'What is 5 multiplied by 4?',
             5: 'Who painted the Mona Lisa?',
             6: 'What gas do plants use for photosynthesis?',
             7: 'In which country is the Great Pyramid of Giza located?',
             8: 'What is the smallest prime number?',
             9: 'Which author wrote “Romeo and Juliet”?',
             10: 'What is the chemical symbol for gold?'}

choices = {1: ['Florida', 'Paris', 'Texas', 'London'],
           2: ['Jupyter', 'Venus', 'Mars', 'Mercury'],
           3: ['Six', 'Eight', 'Four', 'Ten'],
           4: [20, 15, 25, 10],
           5: ['Vincent Van Gogh', 'Pablo Picasso', 'Leonardo da Vinci', 'Claude'],
           6: ['Oxygen', 'Nitrogen', 'Carbon Dioxide', 'Hydrogen'],
           7: ['Mexico', 'India', 'Egypt', 'China'],
           8: [1, 2, 3, 0],
           9: ['Charles Dickens', 'William Shakespeare', 'Jane Austen', 'Mark Twain'],
           10: ['Au', 'Ag', 'Fe', 'Cu']}

answers = {1: 'Paris',
           2: 'Mars',
           3: 'Eight',
           4: 20,
           5: 'Leonardo da Vinci',
           6: 'Carbon Dioxide',
           7: 'Egypt',
           8: 2,
           9: 'William Shakespeare',
           10: 'Au'}

prizes = [100, 200, 300, 500, 1000, 2000, 4000, 8000, 16000, 32000]

def displayQuestion(question, choice):
  print(question, "\n")
  for i in range(1, 5):
    print(f'{i}) {choice[i-1]}')
  print('\n')

def getAnswerAndPrize(question, choice, answer, prize, playerPrize):
  displayQuestion(question, choice)
  ans = int(input("Enter your Answer (1-4): "))
  print("\n")
  while True:
    if ans == 1 or ans == 2 or ans == 3 or ans == 4:
      break
    else:
      ans = int(input("Enter your Answer (1-4): "))
      print("\n")
  if answer == choice[ans-1]:
    print("Congratulations! You're answer is correct")
    playerPrize += prize
    return 1, playerPrize
  else:
    print("[!] You're answer is incorrect.\nBetter Luck next time.\n")
    return 0, playerPrize


def playGame():
  playerPrize = 0
  for i in range(1, 11):
    print(f'***** QUESTION {i} *****')
    question = questions[i]
    choice = choices[i]
    answer = answers[i]
    prize = prizes[i-1]

    continueGame, playerPrize = getAnswerAndPrize(question, choice, answer, prize, playerPrize)

    if continueGame == 1:
      print(f'Accumulated Prize: ${playerPrize}')
      if i != 10:
        print("Let's move on the next question...\n")
      continue
    elif continueGame == 0:
      break
  return playerPrize

print("------------ WHO WANTS TO BE A MILLIONAIRE ------------\n")
playerPrize = playGame()
print(f'\nYour final prize is ${playerPrize}\n')
print("------------ THANKYOU FOR PLAYING THE GAME ------------")
```
