```python
import random

questions = {1: 'What is the capital city of France?',
             2: 'Which planet is known as the Red Planet?',
             3: 'How many legs does a spider typically have?',
             4: 'What is 5 multiplied by 4?',
             5: 'Who painted the Mona Lisa?',
             6: 'What gas do plants use for photosynthesis?',
             7: 'In which country is the Great Pyramid of Giza located?',
             8: 'What is the smallest prime number?',
             9: 'Which author wrote “Romeo and Juliet”?',
             10: 'What is the chemical symbol for gold?'}

choices = {1: ['Florida', 'Paris', 'Texas', 'London'],
           2: ['Jupyter', 'Venus', 'Mars', 'Mercury'],
           3: ['Six', 'Eight', 'Four', 'Ten'],
           4: [20, 15, 25, 10],
           5: ['Vincent Van Gogh', 'Pablo Picasso', 'Leonardo da Vinci', 'Claude'],
           6: ['Oxygen', 'Nitrogen', 'Carbon Dioxide', 'Hydrogen'],
           7: ['Mexico', 'India', 'Egypt', 'China'],
           8: [1, 2, 3, 0],
           9: ['Charles Dickens', 'William Shakespeare', 'Jane Austen', 'Mark Twain'],
           10: ['Au', 'Ag', 'Fe', 'Cu']}

answers = {1: 'Paris',
           2: 'Mars',
           3: 'Eight',
           4: 20,
           5: 'Leonardo da Vinci',
           6: 'Carbon Dioxide',
           7: 'Egypt',
           8: 2,
           9: 'William Shakespeare',
           10: 'Au'}

def chooseRandom(used_questions):
  num = random.randint(1, 10)
  while num in used_questions:
    num = random.randint(1, 10)
  used_questions.append(num)
  return num

def displayQuestionAndChoices(i, num):
  question = questions[num]
  choice = choices[num]
  print(f'***** QUESTION {i} *****')
  print(question, "\n")
  for i in range(1, 5):
    print(f'{i}) {choice[i-1]}')
  print('\n')
  return choice

def validPlayerChoice():
  playerChoice = int(input("Enter your choice (1-4): "))
  print("\n")
  if playerChoice < 1 and playerChoice > 4:
    print("[!] Invalid Choice. Try Again.")
    playerChoice = int(input("Enter your choice (1-4): "))
    print("\n")
  return playerChoice

def playGame():
  used_questions = []
  playerScore = 0
  for i in range(1, 6):
    num = chooseRandom(used_questions)
    answer = answers[num]
    choice = displayQuestionAndChoices(i, num)
    playerChoice = validPlayerChoice()
    if answer == choice[playerChoice-1]:
      print("Congratulations! You're answer is correct")
      playerScore += 1
      print("Score:", playerScore)
      print("\n")
    else:
      print("[!] You're answer is incorrect.")
      print("Score:", playerScore)
      print("\n")
  print("Final Score:", playerScore)

print("------------ WELCOME TO THE QUIZ ------------\n")
playGame()
print("\n")
while True:
  playAgain = input("Do you want to play again? (yes/no): ")
  if playAgain.lower() == 'yes':
    playGame()
  else:
    print("----- Thank you for playing the game! -----")
    break
```
